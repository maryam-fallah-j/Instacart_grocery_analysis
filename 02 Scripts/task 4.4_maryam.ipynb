#01 Importing libraries for Data Wrangling

```python
# import libraries
import pandas as pd
import numpy as np
import os
```

```python
#import orders.csv data set
df_ords = pd.read_csv('orders.csv' , index_col= False)
                    
```

```python
df_ords.head()
```

```python
#02 identifying another variable which is not neccessary in numeric calculations & turning it into a String data type.
df_ords['order_dow'] = df_ords['order_dow'].astype('str')
```

```python
display(df_ords['order_dow'].dtype)
```

```python
#03 renaming an unintutive name of a column in the orders.csv
df_ords.rename(columns = {'order_dow' : 'orders_day_of_week'} , inplace=True)
```

```python
display(df_ords.columns)
```

```python
#04 Finding the busiest hour for placing an order
hour_counts = df_ords['order_hour_of_day'].value_counts()
busiest_hour = hour_counts.idxmax()   # gives the hour (e.g. 14)
```

```python
print('busiest hour is :' , busiest_hour, 'OClock')
```

```python
# importing the second data set: products.csv
df_prods = pd.read_csv ('products.csv', index_col= False)
```

```python
df_prods.head()
```

```python
# importing the third data set: departments.csv
df_dp = pd.read_csv ('departments.csv', index_col= False)
```

```python
display(df_dp.head())
```

```python
#transpositing department data set from wide format to long format
df_dp_t = df_dp.T
```

```python
# we need to adjust the data set's header after transposition. Therefore, the next 3 steps must be concluded:
# 1: make a new header
new_header = df_dp_t.iloc[0]
```

```python
# 2 copy the rest of the data set into a new variable
df_dp_t_new = df_dp_t[1:]
```

```python
# 3 adding the header to the columns
df_dp_t_new.columns = new_header
```

```python
display(df_dp_t_new)
```
